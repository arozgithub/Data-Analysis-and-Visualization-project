```python

# Function to filter out non-English words
def remove_non_english(text):
    # Regular expression to match English words (including contractions)
    english_pattern = re.compile(r'[A-Za-z]+(?:\'[A-Za-z]+)?')
    # Join English words found in the text
    english_words = ' '.join(english_pattern.findall(text))
    return english_words

# Apply the function to the 'title' column
data['title'] = data['title'].apply(remove_non_english)

# Display the updated DataFrame with non-English words removed from the 'title' column
data
```

# Bag of Words(Feature Extraction)

```python
def unique_words(sample):
    words = sample.split()
    unique_words = set(words)
    return list(unique_words)

def bag_of_words_matrix(sample_list):
    word_index = {}
    word_counts = []

    for sample in sample_list:
        if sample is not None:
            words = unique_words(sample)
            doc_counts = [0] * len(word_index)

            for word in words:
                if word not in word_index:
                    word_index[word] = len(word_index)
                    doc_counts.append(1)
                else:
                    word_idx = word_index[word]
                    doc_counts[word_idx] += 1

            word_counts.append(doc_counts)
    columns = word_index
    df_bow = pd.DataFrame(word_counts, columns=columns)
    df_bow=df_bow.fillna(int(0))

    return df_bow
```

```python
titles = data['title']
bow = bag_of_words_matrix(titles)
print(bow)
```

# **Nueral Network CNN(Deep Learning)**

```python
from keras.layers import Embedding, LSTM
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Conv1D, MaxPooling1D, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
# Assuming 'data' contains 'title' and 'category' columns
titles = data['title'].values
categories = data['category'].values

# Tokenize and pad sequences
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(titles)
sequences = tokenizer.texts_to_sequences(titles)
X_pad = pad_sequences(sequences, maxlen=100)

# Encode target variable
encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(categories)

# Splitting data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_pad, Y_encoded, test_size=0.2)

# Model Architecture
model = Sequential()
model.add(Embedding(5000, 128, input_length=100))
model.add(Conv1D(64, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=4))
model.add(LSTM(100))  # Adding LSTM layer
model.add(Dense(len(set(categories)), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=10, batch_size=32, validation_data=(X_test, Y_test))

# Evaluate model accuracy on test set
accuracy_cnn = model.evaluate(X_test, Y_test)
print(f'Test Accuracy: {accuracy_cnn[1]*100:.2f}%')
```

# RNN MODEL

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder



# Preprocessing
max_words = 1000  # Define the maximum number of words
tokenizer = Tokenizer(num_words=max_words, split=' ')
tokenizer.fit_on_texts(data['title'].values)
X = tokenizer.texts_to_sequences(data['title'].values)
X = pad_sequences(X)

encoder = LabelEncoder()
Y = encoder.fit_transform(data['category'])

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Build the model
model = Sequential()
model.add(Embedding(max_words, 128, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(196, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(len(encoder.classes_), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, Y_train, epochs=10, batch_size=32, validation_data=(X_test, Y_test))

# Evaluate the model
score = model.evaluate(X_test, Y_test)
print('Test loss:', score[0])
print('Test accuracy%:', score[1]*100)

```

```python
# Generate predictions on the test set (probabilities)
Y_pred_probs = model.predict(X_test)

# Get the class labels (categories) based on the maximum probability
Y_pred = Y_pred_probs.argmax(axis=1)

# Convert encoded labels back to original categories for generating reports
Y_test_category = encoder.inverse_transform(Y_test)
Y_pred_category = encoder.inverse_transform(Y_pred)

# Generate and print the classification report
Dreport = classification_report(Y_test_category, Y_pred_category)
print("Classification Report:\n", Dreport)

```
